Now Manipulate the data to create training and testing sets

````python
from re import X
# loading Data and Combining
#training_df = pd.read_csv(io.BytesIO(uploaded_training['normalized_dataset_training.csv']))
#testing_df = pd.read_csv(io.BytesIO(uploaded_testing['normalized_dataset_testing.csv']))
# If data is already loaded
training_df = pd.read_csv('normalized_dataset_training.csv')
testing_df = pd.read_csv('normalized_dataset_testing.csv')

combined_data = pd.concat([training_df, testing_df], axis=0)

# Seperating features and Labels (Attack or no attack)
# Split combined data into features and labels
features = combined_data.drop(['label', 'attack_cat'], axis=1).values
feature_names = list(combined_data.columns[:-2])

labels = combined_data['label'].values

# Perform an 80/20 split for training and testing 3 times
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(features, labels, test_size=0.8, random_state=42)
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(features, labels, test_size=0.99, random_state=53)
X_train_3, X_test_3, y_train_3, y_test_3 = train_test_split(features, labels, test_size=0.999, random_state=90)
````

# Train our Model with the different Sets

Parameters:


````python
# Create the Decision Tree
dtree = DecisionTreeClassifier()
# Model 1
dtree_1 = dtree.fit(X_train_1, y_train_1)

# Training Set Accuracy
training_true_pos = 0
training_false_pos = 0
training_true_neg = 0
training_false_neg = 0

y_pred = []
for row in range(len(X_train_1)):
  y_pred.append(dtree.predict([X_train_1[row]]))

correct = 0
for i in range(len(y_pred)):
  if y_pred[i] == y_train_1[i]:
    correct+=1
  # now get the false pos etc
  if y_pred[i] == 1 and y_train_1[i] == 1: # True Positive
    training_true_pos += 1
  elif y_pred[i] == 0 and y_train_1[i] == 1: #
    training_false_neg += 1
  elif y_pred[i] == 0 and y_train_1[i] == 0: #
    training_true_neg += 1
  else: #
    training_false_pos += 1

precision = training_true_pos / (training_true_pos + training_false_pos)
recall = training_true_pos / (training_true_pos + training_false_neg)
F1 = 2 * (precision * recall) / (precision + recall)

print("Number of Training Data Points:" + str(len(y_pred)))
print("Training Set Accuracy: " + str(correct/len(y_pred)))
print("Training F1: " + "{:.2f}".format(F1) + ", Training Precicion: " + "{:.2f}".format(precision) + ", Training Recall: " + "{:.2f}".format(recall))
confusion_matrix = metrics.confusion_matrix(y_train_1, y_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

cm_display.plot()
plt.show()
print("\n")
# Now, Testing Set Accuracy
testing_true_pos = 0
testing_false_pos = 0
testing_true_neg = 0
testing_false_neg = 0

y_pred = []
for row in range(len(y_test_1)):
  y_pred.append(dtree.predict([X_test_1[row]]))

correct = 0
for i in range(len(y_pred)):
  if y_pred[i] == y_test_1[i]:
    correct+=1
  # now get the false pos
  if y_pred[i] == 1 and y_test_1[i] == 1: # True Positive
    testing_true_pos += 1
  elif y_pred[i] == 0 and y_test_1[i] == 1: #
    testing_false_neg += 1
  elif y_pred[i] == 0 and y_test_1[i] == 0: #
    testing_true_neg += 1
  else: #
    testing_false_pos += 1

precision = testing_true_pos / (testing_true_pos + testing_false_pos)
recall = testing_true_pos / (testing_true_pos + testing_false_neg)
F1 = 2 * (precision * recall) / (precision + recall)

print("Number of Testing Data Points:" + str(len(y_pred)))
print("Testing Set Accuracy: " + str(correct/len(y_pred)))
print("Training F1: " + "{:.2f}".format(F1) + ", Training Precicion: " + "{:.2f}".format(precision) + ", Training Recall: " + "{:.2f}".format(recall))

confusion_matrix = metrics.confusion_matrix(y_test_1, y_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

cm_display.plot()
plt.show()
````
